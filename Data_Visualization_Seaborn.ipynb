# SEABORN

- Seaborn is used for Data Visualization
- Seaborn is widely used for statistical plottings and Distributions.
- It has some inbuilt reserch datasets as well that we can use.

```python
import pandas as pd
import numpy as np
import seaborn as sns  # SNS will known as SEABORN
import matplotlib.pyplot as plt
```

```python
# Seaborn has some reserch dataset inbuilt.

# If we want to see those dataset names:

sns.get_dataset_names()  # With this method we will see the names of datasets.
```

```python
# How to lead dataset and work with them in seaborn.

df = sns.load_dataset('diamonds')
```

```python
# Now df contains the Diamonds Dataset.

df
```

```python
# Basic Plotting in Matoplotlib with Lineplot,barplot,pie,scatter.

# Seaborn is used the most for Data Science purpose as it provide data distributions.

# LINEPLOT in SEABORN.

# As we know line plot takes X & Y as Numeric (Continious Numeric)
# In Matplotlib we did,'t descrive or mention x and y we directly pass the column name, but in seaborn we've to pass an argument s as x and y.


sns.lineplot(x = df.depth,y = df.price)

# In seaborn we' don't to give labels as it picks the column name on its own.
```

```python
# To resize the figure in seaborn we still use MATPLOTLIB.

plt.figure(figsize=(8,4))
sns.lineplot(x = df.depth,y = df.price)
```

```python
# All the arguments and Parameters will remains same.

plt.figure(figsize=(11,4))
sns.lineplot(x = df.depth,y = df.price, color='r')
```

```python
# There is a very important feature or argument in seaborn named as HUE.
# HUE will diffrentiate the data (x and y) w.r.t any other categorial column.

plt.figure(figsize=(11,4))
sns.lineplot(x = df.depth,y = df.price, color='r',hue=df.cut)
```

```python
# Scatter PLot

# Scatter plot will show us the spread of Data.

plt.figure(figsize=(11,4))
sns.scatterplot(x=df.depth,y=df.table,hue=df.color)
```

```python
# Barplot
# Barplot take categorical names as X axis and Their counts as Y Axis.
# In Seaborn each category has described ot visualize with different color.

plt.figure(figsize=(11,4))
sns.barplot(x=df.cut.value_counts().index,y=df.cut.value_counts())
```

```python
# How to plot that kind of data easily with countplot.
# In New Version we use x (sns.countplot(x=df.cut))
# We can also diffrentate this data with Hue


sns.countplot(df.cut,hue=df.color)
```

```python
# Matplotlib is used to create data distributaions as well but mathematically puposes:
# HISTOGRAMS
# Data Distribution can only work with Single column at a time.
# Passed data has to be continious numeric in nature.

plt.hist(df.price)
```

```python
# Mathematical Histogram won't allow us to visualize the distribution better.

# For distribution we use histograms from SEABORN.

sns.displot(df.price)
```

```python
# We can use DistPLot as well that will show us the curves with histogram.

sns.distplot(df.price)
```

```python
# Math will work on COUNTS and Stats will work on DENSITY.

# We can use some arguments with it as well.
# if we want to replace or remove that curve use use KDE arguments.

plt.figure(figsize=(8,4))
sns.distplot(df.price,kde=False,color='green')
plt.title('Data Distribution for DIAMONDS PRICES')
```

```python
# We can also use the KDE PLOT to check for the curved distributions

# KDE PLOT - - - >>> Kernel Distribution Estimator

sns.kdeplot(df.price,fill=True
           )
```

```python
# We can also diffrentiate the data with HUE in those Curves.

sns.kdeplot(df.price,hue=df.cut,fill=True)
```

```python
# PAir Plot

# Pair plot will show the scatterness and distributions of each column w.r.t to other column individually.

sns.pairplot(df)
```

```python
# Check for the Correlation between the columns

df.corr()
```

```python
# How do we visualize the correlation plot.

sns.heatmap(df.corr())
```

```python
# We can use correlation and plot thatt text on this heatmap.

sns.heatmap(df.corr(),annot=True,cmap='Reds')
```
