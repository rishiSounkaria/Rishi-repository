# MATPLOTLIB (Math Plotting)

```python
# How to import thsi library matplotlib ?
# We use a specific library sub- module named as pyplot from matplotlib.

from matplotlib import pyplot as plt
```

```python
import matplotlib.pyplot as plt
```

```python
# LINEPLOT

# It has two axis x & y.
# Both the axis take numeric data not categorical data.
```

```python
# First we'll create an array in X and Y then we'll plot them in a Line PLot

x = np.array([1,2,3,4,5])
y = np.array([10,20,30,40,50])
```

```python
plt.plot(x,y)
```

```python
# If we plot a line plot there will be no data point shown.
# For Data Points we can use an argument named as marker

plt.plot(x,y,marker='p',ms=14)  # There are other markers as well that we can use.
```

```python
# Let's import MAgic Bricks Dataset 

df = pd.read_csv('house pricing delhi- MagicBricks.csv')
```

```python
df
```

```python
# Let's try to create Line PLot with Two Numeric(Continious Numeric) column on both the AXIS (X & Y).

plt.plot(df.Area,df.Price,marker='o')
```

```python
# If there is so much data for visualization we can use head method in those cases.

plt.plot(df.Area.head(10),df.Price.head(10),marker='h')
```

```python
# We can also use some other arguments or p[arameters like markersize and colors.

plt.plot(df.Area.head(10),df.Price.head(10),marker='h',color='#349F0C',markersize=12)
```

```python
# How to Resize the plot accordingly.

plt.figure(figsize=(18,5))    # (Width,Length)
plt.plot(df.Area.head(10),df.Price.head(10),marker='h',color='#349F0C',markersize=12)
```

```python
# As we are using Matplotlib there will be no axis name given by default we've to provide axis name.

plt.figure(figsize=(18,5))    # (Width,Length)
plt.plot(df.Area.head(10),df.Price.head(10),marker='h',color='#349F0C',markersize=12)
plt.xlabel('House Area')
plt.ylabel('House Price')
plt.title('Delhi House Price and Area')
plt.grid(color='r')
```

```python
# We can also incre the line width and Line Style

plt.figure(figsize=(18,5))    # (Width,Length)
plt.plot(df.Area.head(10),df.Price.head(10),marker='h',color='#349F0C',markersize=12,linewidth=4.5,linestyle='-.')
plt.xlabel('House Area')
plt.ylabel('House Price')
plt.title('Delhi House Price and Area')
plt.grid(color='r')
```

```python
# Scatterplot is used see the scatterness of Data. 
# In Scatter PLot both x and Y has to be continiooous Numerc .


plt.scatter(df.Area,df.Price)
```

```python
plt.figure(figsize=(18,5))    # (Width,Length)
plt.scatter(df.Area,df.Price,color='#349F0C')
plt.xlabel('House Area')
plt.ylabel('House Price')
plt.title('Delhi House Price and Area')
plt.grid()
```
